```python
from torchvision import transforms as T
from PIL import Image
```

```python
tilef = '/fast/rsna-breast/topTiles/4612225_7_2.png'
tile = Image.open(tilef).convert('RGB')
tile
```

```python
T.GaussianBlur((1,1), sigma=0.01)(tile)
```

```python
T.ColorJitter(hue=0.5)(tile)
```

```python
T.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)(tile)
```

```python
T.RandomAffine(90,shear=(-20.0,20.0), scale=(0.5,1.5))(tile)
```
